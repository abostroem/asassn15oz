```python
import os

from matplotlib import pyplot as plt
from astropy.io import ascii as asc
import astropy.units as u
from astropy.time import Time

import supernova as sn
%matplotlib inline
```

```python
sn15oz = sn.LightCurve2('asassn15oz')
```

```python
sn15oz.get_photometry('V')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(sn15oz.jd['V'] - sn15oz.jdexpl, sn15oz.apparent_mag['V'], 'o')
ax.set_ylim(ax.get_ylim()[::-1])
ax.set_xlabel('Epoch (days since explosion)')
ax.set_ylabel('Apparent Magnitude')
ax.set_title('V band Apparent Magnitude')
```

```python
sn15oz.get_abs_mag('V')
```

```python
mag_table = asc.read('../../not_my_code/SNEC-1.01/Data/magnitudes.dat', data_start=0)
```

```python
fig = plt.figure(figsize=[11,8])
ax = fig.add_subplot(111)
ax.plot(sn15oz.jd['V'] - (sn15oz.jd['V'][0]-10), sn15oz.abs_mag['V'], 'o', label='V Band Data')
ax.plot((mag_table['col1']*u.second).to(u.day), mag_table['col11'], label='Model')
ax.set_ylim(ax.get_ylim()[::-1])
ax.set_xlabel('Epoch (days since explosion)')
ax.set_ylabel('Absolute Magnitude')
ax.set_title('V band Light Curve for ASASSN-15oz')
plt.legend(loc='best')
plt.savefig('../figures/test_lc_model_comp.pdf')
```
