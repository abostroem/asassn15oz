```python
import visualization as viz
import numpy as np
from matplotlib import pyplot
import glob
import os
from astropy.io import fits
```

```python
%matplotlib inline
```

```python
DATA_DIR = '../data/spectra/lco'
FIG_DIR = '../figures'
```

```python
flist = glob.glob(os.path.join(DATA_DIR, '*multi.fits'))
for i, ifile in enumerate(flist):
    if '105042.698_multi.fits' in ifile:
        flist.pop(i)
        print('removing', ifile)
flist = np.array(flist)
date_obs = []
for ifile in flist:
    date_obs.append(fits.getval(ifile, 'date-obs', 0))
date_obs = np.array(date_obs)
```

```python
indx = np.argsort(date_obs)
fig = pyplot.figure(figsize=[20, 30])
ax = fig.add_subplot(1,1,1)
viz.plot_multiple_spectra(flist[indx], ycol='OPTIMAL_FLUX', separation = 0.8, ax=ax)
leg_list = [idate.split('T')[0] for idate in date_obs[indx]]
pyplot.legend(leg_list, loc='best', framealpha=0.25, ncol=4)
pyplot.title('ASAS-SN15oz LCO Spectra')
pyplot.xlabel('Wavelength')
pyplot.ylabel('Flux')
pyplot.savefig(os.path.join(FIG_DIR, 'lco_spectra.pdf'))
```
